```python
'''
Data:2018-10-30
Author: Kai Zhang
Stevens Intitute of Technology
Description:merge multiple files into a single csv file
'''
import pandas as pd
import os
Folder_Path = r'C:\Users\yongk\Documents\PythonLearning\Steam Data Analysis\Data'          
#要拼接的文件夹及其完整路径，注意不要包含中文
SaveFile_Path =  r'C:\Users\yongk\Documents\PythonLearning\Steam Data Analysis\Data'       
#拼接后要保存的文件路径
SaveFile_Name = r'game_data_all.csv'              
#合并后要保存的文件名
 
#修改当前工作目录
os.chdir(Folder_Path)
#将该文件夹下的所有文件名存入一个列表
file_list = os.listdir()
 
#读取第一个CSV文件并包含表头
df = pd.read_csv(Folder_Path +'\\'+ file_list[0])   #编码默认UTF-8，若乱码自行更改
 
#将读取的第一个CSV文件写入合并后的文件保存
df.to_csv(SaveFile_Path+'\\'+ SaveFile_Name,encoding="utf_8_sig",index=False)
 
#循环遍历列表中各个CSV文件名，并追加到合并后的文件
for i in range(1,len(file_list)):
    df = pd.read_csv(Folder_Path + '\\'+ file_list[i])
    df.to_csv(SaveFile_Path+'\\'+ SaveFile_Name,encoding="utf_8_sig",index=False, header=False, mode='a+')
```

```python
df = pd.read_csv('C:/Users/yongk/Documents/PythonLearning/Steam Data Analysis/Data/game_data_all.csv')
# remove the unused column which is generated when merging multiple files
df=df.drop(['Unnamed: 0'], axis=1)
# sorting by first name 
df.sort_values("names", inplace = True) 
# dropping ALL duplicte values 
df.drop_duplicates(subset ="names", 
                     keep = False, inplace = True) 
```

```python
month_lst = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
              'August', 'September', 'October', 'November', 'December']
```

```python
len(df)
```

```python
date_lst = df['post_date'].tolist()
date_lst
```

```python
date_lst[0].split(' ')[0]
```

```python
def monthToNum(shortMonth):
    return{
        'January' : 1,
        'February' : 2,
        'March' : 3,
        'April' : 4,
        'May' : 5,
        'June' : 6,
        'July' : 7,
        'August' : 8,
        'September' : 9, 
        'October' : 10,
        'November' : 11,
        'December' : 12
    }[shortMonth]
```

```python
new_date_lst=[]
for i in range(1,len(date_lst)):
    if date_lst[i].split(' ')[0] in month_lst:
        new_date_lst.append(monthToNum(date_lst[i].split(' ')[0]))
new_date_lst
        
```
